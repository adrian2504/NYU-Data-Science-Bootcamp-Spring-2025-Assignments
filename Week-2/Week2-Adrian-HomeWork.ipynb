````python
import numpy as np
````

````python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
iris_2d = np.genfromtxt(url, delimiter=',', dtype='float', usecols=[0,1,2,3])
````

````python
iris_2d
````

````python
# 1

A = np.array([1, 2, 5, 8, 10, 12])
B = np.array([3, 5, 8, 11, 18, 20])

# Vertical stacking: each array becomes a row in a new 2D array.
vertical_stack = np.vstack((A, B))
print("Vertical Stack: \n", vertical_stack)

horizonal_stack = np.hstack((A,B))
print("\n Horizontal Stack \n", horizonal_stack)
````

````python
#2

common_ele = np.intersect1d(A,B)
print(" \n Common elements are: \n", common_ele)
````

````python
# 3

number_range = (A>=5) & (A<=10)
print(" numbers in array A in range from 5 to 10 are \n", A[number_range])
````

````python
iris_val = iris_2d[(iris_2d[:,2] > 1.5) & (iris_2d[:, 0] < 5.0) ]
print("The array is: ", iris_val)
````
